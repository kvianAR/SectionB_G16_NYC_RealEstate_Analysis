## 4. Cleaning Pipeline

**Working copy** — original `df_raw` is never modified.

```python
df = df_raw.copy()
```

### 4.1 Drop Irrelevant Columns

`APARTMENT NUMBER`, `EASE-MENT`, and `Unnamed: 0` carry no analytical value.

```python
df = df.drop(columns=["APARTMENT NUMBER", "EASE-MENT", "Unnamed: 0"], errors="ignore")
print(f"Columns after drop: {df.shape[1]}")
print(df.columns.tolist())
```

### 4.2 Standardise Column Names

Lowercase, strip whitespace, replace spaces with underscores.

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
print(df.columns.tolist())
```

### 4.3 Fix Data Types

Numeric columns are stored as strings with commas or dash placeholders. Date columns need parsing.

```python
for col in ["sale_price", "gross_square_feet", "land_square_feet"]:
    df[col] = (
        df[col].astype(str)
        .str.replace(",", "")
        .str.strip()
        .replace("-", np.nan)
    )
    df[col] = pd.to_numeric(df[col], errors="coerce")

df["sale_date"] = pd.to_datetime(df["sale_date"], errors="coerce")
df["year_built"] = pd.to_numeric(df["year_built"], errors="coerce")

print("Data types after conversion:")
print(df.dtypes)
```

### 4.4 Fix `tax_class_at_present` — Hidden Missing Values

Whitespace strings masquerade as valid entries. Strip, replace with NaN, and drop.

```python
print(f"Rows before: {len(df)}")

df["tax_class_at_present"] = df["tax_class_at_present"].astype(str).str.strip()
df["tax_class_at_present"] = df["tax_class_at_present"].replace(["", "nan"], np.nan)
df = df.dropna(subset=["tax_class_at_present"])
df["tax_class_at_present"] = df["tax_class_at_present"].astype("category")

print(f"Rows after: {len(df)}")
print(f"Valid categories: {df['tax_class_at_present'].cat.categories.tolist()}")
```

### 4.5 Fix `year_built` — Zero as Missing

Year 0 is not a valid build year. Replace with borough-level median.

```python
print(f"Zero year_built entries: {(df['year_built'] == 0).sum()}")
df["year_built"] = df["year_built"].replace(0, np.nan)
df["year_built"] = df.groupby("borough")["year_built"].transform(lambda x: x.fillna(x.median()))
print(f"Remaining nulls in year_built: {df['year_built'].isnull().sum()}")
```

### 4.6 Remove Non-Market Sales

Sales below $10,000 are intra-family transfers or administrative entries — not arm's-length market transactions. Industry standard threshold applied.

```python
print(f"Zero sale prices: {(df['sale_price'] == 0).sum()}")
print(f"Sale prices < $10,000: {(df['sale_price'] < 10000).sum()}")
print(f"Null sale prices: {df['sale_price'].isnull().sum()}")

df = df[df["sale_price"] >= 10000]
print(f"\nRows after removing non-market sales: {len(df)}")
```

### 4.7 Handle Missing Square Footage

Rows with no gross square footage cannot be used for price-per-sqft KPIs. These represent less than 5% of the dataset.

```python
print(f"Null gross_square_feet: {df['gross_square_feet'].isnull().sum()}")
print(f"Null land_square_feet: {df['land_square_feet'].isnull().sum()}")

df = df.dropna(subset=["gross_square_feet", "land_square_feet"])
print(f"Rows after dropping null sqft: {len(df)}")
```

### 4.8 Remove Outliers — Square Footage (IQR Method)

Properties with zero or statistically extreme sizes are removed. IQR fencing retains 99%+ of legitimate properties.

```python
print(f"Zero sqft rows: {(df['gross_square_feet'] == 0).sum()}")
df = df[df["gross_square_feet"] > 0]

Q1 = df["gross_square_feet"].quantile(0.25)
Q3 = df["gross_square_feet"].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

print(f"Lower fence: {lower:.0f} sqft")
print(f"Upper fence: {upper:.0f} sqft")
print(f"Rows before: {len(df)}")

df = df[(df["gross_square_feet"] >= lower) & (df["gross_square_feet"] <= upper)]
print(f"Rows after: {len(df)}")
```

### 4.9 Feature Engineering

Derived columns that are used in EDA and KPI analysis notebooks.

```python
# Building age
df["building_age"] = df["sale_date"].dt.year - df["year_built"]

# Core KPI: Price per square foot
df["price_per_sqft"] = df["sale_price"] / df["gross_square_feet"]

# Time-series features
df["sale_year"]  = df["sale_date"].dt.year
df["sale_month"] = df["sale_date"].dt.month

print("Derived columns added: building_age, price_per_sqft, sale_year, sale_month")
df[["building_age", "price_per_sqft", "sale_year", "sale_month"]].describe()
```

## 5. Final Quality Check

```python
print("===== FINAL CLEANED DATASET ======")
print(f"Shape: {df.shape}")
print(f"\nNull values:")
print(df.isnull().sum())
print(f"\nDuplicates: {df.duplicated().sum()}")
print(f"\nSale price range: ${df['sale_price'].min():,.0f} — ${df['sale_price'].max():,.0f}")
print(f"Year range: {int(df['sale_year'].min())} — {int(df['sale_year'].max())}")
df.info()
```
